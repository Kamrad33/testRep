```python
# O(n)  
# n = 2 =>  1
# n = 3 =>  2
# n = 4 =>  3 
def fib(n):
    fibarr=[]
    if n<=1:
        for i in range(n+1):
            fibarr.append(i)
            print (fibarr)
    else:
        fibarr=[0,1]
        for i in range(1, n):
            num= fibarr[i-1]+fibarr[i]
            fibarr.append(num)
            print (fibarr)
            print ("fib:", fibarr[i+1])
            
            
fib(100)
```

```python
#   +- 50000 
def recFib(n: int):
    if n <=1:
        return n
    else:
        return recFib(n-1)+recFib(n-2)

print(recFib(3))
print(recFib(5))
print(recFib(10))
print(recFib(100))
```

```python
# 
arr = [num for num in range(-10, 10)]
print (arr)
def binarySearch(arr, key):
    low = 0
    high = len(arr)-1
    while low <=high:
        mid = low + (high-low)//2
        if key < arr[mid]:
            high = mid-1
        elif key > arr[mid]:
            low = mid+1
        else:
            return mid           
        
    return -1
print(binarySearch(arr, 1))
```
